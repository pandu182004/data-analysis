```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\kanuk\OneDrive\Desktop\cognify\Dataset .csv")
print(df.columns)
```

4.1 Determine the percentage of restaurants
that offer online delivery.

```python
online_delivery=df['Has Online delivery'].value_counts().get('Yes',0)
print(online_delivery)
no=df['Has Online delivery'].value_counts().get('No',0)
print("total restuarants with no delivery:",no)
row_count=len(df)
percentage_of_online_delivery=round((online_delivery/row_count)*100,2)
print("Total % of online delivery:",percentage_of_online_delivery)
```

4.2 Compare the average ratings of restaurants
with and without online delivery.

```python

avg_rating_yes = round(df[df["Has Online delivery"] == 'Yes']['Aggregate rating'].mean(), 2)
print("Average rating with online delivery (Yes):", avg_rating_yes)
avg_rating_no = round(df[df["Has Online delivery"] == 'No']['Aggregate rating'].mean(), 2)
print("Average rating with online delivery (No):", avg_rating_no)
```

```python
labels=['YES','NO']
avg_rating=[avg_rating_yes,avg_rating_no]
colors='lightgreen','pink'
plt.bar(labels,avg_rating,color=colors)
plt.ylabel('online_delivery')
plt.xlabel('avg_rating')
plt.title('Comparision of average ratings of online delivery')
plt.show()
```
