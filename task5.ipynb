Task: Restaurant Ratings

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\kanuk\OneDrive\Desktop\cognify\Dataset .csv")
agg_rating_distribution=df['Aggregate rating'].value_counts()
print(agg_rating_distribution)
```

1.1 Analyze the distribution of aggregate ratings and determine the most common rating range

```python
m=agg_rating_distribution.idxmax()
print("Most common rating: ",m)
plt.hist(df['Aggregate rating'],bins=30,color='lightgreen',edgecolor='black')
plt.xlabel('Aggregate rating')
plt.ylabel('Count')
plt.title("Distribution of aggregate rating")
plt.show()
```

1.2 Calculate the average number of votes
received by restaurants.

```python
avg_vote=round(df['Votes'].mean(),3)
print("The avearge no of votes received by restaurants: ",avg_vote)
```
