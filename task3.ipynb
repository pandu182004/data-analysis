```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:/Users/kanuk/OneDrive/Desktop/cognify/Dataset .csv")
df.head()
```

3.1 Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.

```python
df.hist(column="Price range")
plt.title("distribution of price ranges")
plt.xlabel('price range')
plt.ylabel("counts")
plt.show()
```

Calculate the percentage of restaurants each price range category


```python
price_range_count=df["Price range"].value_counts()
total_restaurants=len(df)
percentage = round((price_range_count/total_restaurants)*100,2)
percentage_m=pd.merge(price_range_count,percentage,on="Price range")
print("the % of restaurants in each Price range category are: ",percentage_m)
```

```python
plt.pie(percentage,labels=percentage.index,autopct='%1.1f%%',colors=["red","blue","white","green"])
plt.title("Price range percentage")
plt.show()
```
